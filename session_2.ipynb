```python
# =============================
# Student Performance System
# =============================

# Global data storage
students = []        # list of dictionaries
passed_students = set()
failed_students = set()

print("Welcome to Student Performance System")

while True:
    print("\n1- Add Student")
    print("2- View Report")
    print("3- Exit")

    choice = input("Enter your choice: ")

    # ---------- match statement ----------
    match choice:
        case "1":
            name = input("Enter student name: ")

            marks = []
            for i in range(3):
                mark = float(input(f"Enter mark {i+1}: "))
                marks.append(mark)

            # ---------- Arithmetic operators ----------
            total = marks[0] + marks[1] + marks[2]
            average = total / 3

            # ---------- Comparison & Logical operators ----------
            if average >= 50 and all(m >= 40 for m in marks):
                status = "Pass"
                passed_students.add(name)
            else:
                status = "Fail"
                failed_students.add(name)

            # ---------- Dictionary ----------
            student = {
                "name": name,
                "marks": marks,
                "average": average,
                "status": status
            }

            students.append(student)
            print(f"{name} added successfully!")

        case "2":
            if len(students) == 0:
                print("No students found.")
                continue

            print("\n----- Student Report -----")

            # ---------- for loop ----------
            for student in students:
                print("Name:", student["name"])
                print("Marks:", student["marks"])
                print("Average:", student["average"])

                # ---------- if conditional ----------
                if student["status"] == "Pass":
                    print("Status: ✅ Passed")
                else:
                    print("Status: ❌ Failed")

                print("-" * 20)

            # ---------- Non-primitive types ----------
            print("Passed students:", passed_students)
            print("Failed students:", failed_students)

        case "3":
            print("Exiting program...")
            break

        case _:
            print("Invalid choice, try again!")
```

## Sets

```python
basket = {'apple', 'orange', 'apple', 'pear', 'orange', 'banana'}
print(basket)                      # show that duplicates have been removed

print('orange' in basket)                 # fast membership testing

print('crabgrass' in basket)


# Demonstrate set operations on unique letters from two words

a = set('abracadabra')
b = set('alacazam')
print(a)                                  # unique letters in a

print(a - b)                              # letters in a but not in b

print(a | b)                              # letters in a or b or both

print(a & b)                              # letters in both a and b

print(a ^ b)                              # letters in a or b but not both
```

```python
lst = [1, 2, 3]
tpl = (1, 2, 3)
st = {1, 2, 3}
dct = {'a': 1, 'b': 2}

print(len(lst))  # 3
print(len(tpl))  # 3
print(len(st))   # 3
print(len(dct))  # 2 (counts keys)
```

```python
print(2 in lst)      # True
print(2 in tpl)      # True
print(2 in st)       # True
print('a' in dct)    # True (checks keys)
```

```python
print("List:")
for x in lst:
    print(x)

print("Tuple:")
for x in tpl:
    print(x)

print("Set:")
for x in st:
    print(x)

print("Dict:")
for x in dct:  # iterates keys
    print(x)
```

```python
lst2 = [0, 0, 1]
tpl2 = (0, 0, 1)
st2 = {0, 0, 1}
dct2 = {0: 'a', 1: 'b'}  # keys used for truth test

print(any(lst2))  # True
print(any(tpl2))  # True
print(any(st2))   # True
print(any(dct2))  # True (1 key is truthy)
```
